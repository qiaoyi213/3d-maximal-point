```python
import matplotlib.pyplot as plt
```

```python
N = 20000
```

```python
fa = open('a.txt', "r")
fb = open('b.txt', "r")
fc = open('c.txt', "r")
a = []
b = []
c = []
for i in range(N):
    a.append(float(fa.readline()))
    b.append(float(fb.readline()))
    c.append(float(fc.readline()))
```

```python
x = range(1,N+1)
```

```python

plt.rcParams["font.sans-serif"]=["Arial Unicode MS"] 
plt.plot(x,a, label="")
plt.plot(x,b, label="")
plt.plot(x,c, label="")
plt.title(" Maximal point ")
plt.legend()
plt.xlabel(" N")
plt.ylabel("")
plt.show()
```
