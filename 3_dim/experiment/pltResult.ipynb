```python
import matplotlib.pyplot as plt
```

```python
N = 20000
```

```python
fa = open('a.txt', "r")
fb = open('b.txt', "r")
fc = open('c.txt', "r")
a = []
b = []
c = []
for i in range(N):
    a.append(float(fa.readline()))
    b.append(float(fb.readline()))
    c.append(float(fc.readline()))
```

```python
x = range(1,N+1)
```

```python

plt.rcParams["font.sans-serif"]=["Arial Unicode MS"] 
plt.plot(x,a, label="暴力求解")
plt.plot(x,b, label="剪枝")
plt.plot(x,c, label="分治")
plt.title("三維 Maximal point 時間複雜度比較")
plt.legend()
plt.xlabel("每筆資料中點的數量 N（單位：個）")
plt.ylabel("執行時間（單位：秒）")
plt.show()
```
